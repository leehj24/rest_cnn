```python
# seaborn/scipy 문제 회피: 기본 matplotlib 표로 다시 시각화
import matplotlib.pyplot as plt
import pandas as pd

# 정량 지표 예시 값
data = {
    "Metric": ["Accuracy", "R² Score", "mAP@50"],
    "Score": [0.942, 0.891, 0.768]
}

df = pd.DataFrame(data)

# 시각화
fig, ax = plt.subplots(figsize=(6, 3))
ax.bar(df["Metric"], df["Score"], color="red")
ax.set_ylim(0, 1.05)
ax.set_ylabel("Score")
ax.set_title("Segmentation Performance Metrics")

# 값 표기
for i, v in enumerate(df["Score"]):
    ax.text(i, v + 0.02, f"{v:.3f}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
img_path = "./segmentation_metrics_chart.png"
plt.savefig(img_path)
plt.close()

img_path
```
